## Troubleshooting

| Problem | Fix |
|---------|-----|
| `FileNotFoundError: Socratic_OT folder not found` | Upload `Socratic_OT` to Google Drive at `My Drive/Socratic_OT/` |
| `GROQ_API_KEY not set` | Add it in Colab Secrets (🔑) or paste in Cell 2 |
| `ModuleNotFoundError: chromadb` | Re-run Cell 1 |
| `CUDA out of memory` (LLaVA) | Go to Runtime → Change runtime type → T4 GPU |
| Gradio link expired | Re-run Cell 3 |
| Chunks file not found | Make sure the file is named exactly `text_chunks_full 2.csv` (with space) |
